```python
import bs4 as bs
import urllib.request
```

```python
# open a URL with urlopen method and get the source code
sauce = urllib.request.urlopen("https://listiyo.com/").read()
```

```python
# give a (beautiful) structure to the document
# Arguments of BeutifulSoup() method:
# first is the document got from the URL, second is how we want to shape the code
soup = bs.BeautifulSoup(sauce, 'lxml')
```

```python
print(soup)
```

```python
# as a second argument to BeautifulSoup() method we can use html.parser
soup = bs.BeautifulSoup(sauce, 'html.parser')

# then print the result by making it pretty with prettify() method
print(soup.prettify())
```

```python
# get tags of the HTML
print(soup.title)
```

```python
# print the content of given tag
print(soup.title.string)
```

```python
# get first p tag
print(soup.p)
```

```python
# find all p tag and make a list with them.
print(soup.find_all('p'))
```

```python
# find all a tags and print their href attribute if not None
for link in soup.find_all('a'):
    if link.get('href') != None:
        print(link.get('href'))
```

```python
# parse string inside the given tag
for link in soup.find_all('p'):
    print(link.string)
```

```python
# parse string inside the given tag
for link in soup.find_all('p'):
    print(link.text)
```

```python
# not a very efficient method but ...
print(soup.get_text())
```
